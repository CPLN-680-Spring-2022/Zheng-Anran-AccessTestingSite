```python
import pandas as pd
import csv
import numpy as np
```

```python
walk_origin2=pd.read_csv("origin_revise.csv")

wk_line=pd.read_csv("wk_line.csv")
```

```python
wk_line.head()
```

```python
walk_origin=walk_origin2[['ID','TotalPop']]

walk_origin.set_index('ID',inplace=True)
```

```python
walk_origin2.head()
```

```python
walk_merge=pd.merge(wk_line,walk_origin,left_on='OriginID',right_index=True,how='left',sort=False)

walk_merge
```

```python
walk_merge['access']=walk_merge.time_daosh*walk_merge.TotalPop
walk_group=walk_merge.groupby('OriginID')
walk_access=walk_group['access'].sum()
walk_access
```

```python
walk_access1=walk_access.to_frame()
walk_access1
```

```python
#final output
walk_access1.reset_index()
```

```python
walk_origin2
```

```python
walk_origin1=walk_origin2[['ID','TotalPop']]
walk_origin1=walk_origin1.rename(columns={'ID':'OriginID'})
walk_origin1
```

```python
walk_origin3=walk_origin1.merge(walk_access1,on='OriginID',how='left')
walk_origin3['access']=walk_origin3['access'].replace(np.nan, 0)
```

```python
walk_origin3
```

```python
walk_origin3.to_csv("walk_access.csv")
```
