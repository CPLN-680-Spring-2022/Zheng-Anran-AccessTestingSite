```python
primary0=car[(car["type"] == 'primary') | (car["type"] == 'primary_link')]
```

```python
primary0['maxspeed'].replace(0,40,inplace=True)
```

```python
primary0
```

```python
motorway0=car[(car["type"] == 'motorway') | (car["type"] == 'motorway_link')]
```

```python
motorway0['maxspeed'].replace(0,50,inplace=True)
```

```python
motorway0
```

```python
secondary0=car[(car["type"] == 'secondary') | (car["type"] == 'secondary_link')]
secondary0['maxspeed'].replace(0,30,inplace=True)
secondary0
```

```python
tertiary0=car[(car["type"] == 'tertiary') | (car["type"] == 'road')]
tertiary0['maxspeed'].replace(0,25,inplace=True)
tertiary0
```

```python
residential0=car[(car["type"] == 'residential') | (car["type"] == 'service')| (car["type"] == 'services')]
residential0['maxspeed'].replace(0,15,inplace=True)
residential0
```

```python
dfs=[primary0,motorway0,residential0,tertiary0,secondary0]

car0=pd.DataFrame()

car0=pd.concat(dfs)
```

```python
car0
```

```python
car0['spd_kmperh']=car0['maxspeed']*1.6
```

```python
car0['time']=60*car0['length']/car0['spd_kmperh']
car0['time']
```

```python
car0.sort_values(by=['OBJECTID'], ascending=True)
```

```python
car0.to_csv("car0.csv")
```
