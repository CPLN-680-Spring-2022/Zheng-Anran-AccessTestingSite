```python
import pandas as pd
import jenkspy
```

```python
# -*- coding: utf-8 -*-
# @Desc    : natural break jenk classification of influencing factors

excel_path = r"final_ZCTA.xlsx"
class_type = 5 
out_path = r"NYC_break.xlsx"

ret = {
    }
df = pd.read_excel(excel_path)
for col_name in df:
    col = df[col_name]
    breaks = jenkspy.jenks_breaks(col, nb_class=class_type) 
    ret[col_name] = breaks

out_df = pd.DataFrame(ret)
#out_df.to_excel(out_path)
```

```python
out_df
```

```python
pop_den=out_df['pop_den'].tolist()
MedInc=out_df['MedInc'].tolist()
MedAge=out_df['MedAge'].tolist()
Per_White=out_df['Per_White'].tolist()
Per_Bach=out_df['Per_Bach'].tolist()
Per_Public=out_df['Per_Public'].tolist()
positive=out_df['PERCENT_PO'].tolist()
hosp_den=out_df['hosp_den'].tolist()
transit_den=out_df['transit_den'].tolist()
non_transit_den=out_df['non_transit_den'].tolist()
stop_den=out_df['stop_den'].tolist()
```
